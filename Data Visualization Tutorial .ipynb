### INTRODUCTION TO DATA VISUALSATION

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
a = [0,1,2,3,4,5]
b = [0,2,4,6,8,10]

#Resize the graph
plt.figure(figsize=(8,5), dpi=300)

plt.plot(a, b, label='2x', color = 'red', linewidth =2, 
        marker = 'o', markersize = '5', linestyle='solid', markeredgecolor ='k')

#line two

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:5], x2[:5]**2, color= 'blue', label ='X^2')
plt.plot(x2[4:], x2[4:]**2, 'b--')

plt.title('Our First Graph', fontdict={'fontname':'Comic Sans MS', 'fontsize': 20})
plt.xlabel('A axis')
plt.ylabel('B axis')

# plot(x, y, 'go--', linewidth=2, markersize=12) # shot hand notation
# plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)

#fmt = '[color][marker][line]'

plt.legend()
plt.xticks(a)
# plt.yticks(b)

# saving the graph
plt.savefig('Our_Graph')

plt.show()
```

## Bar Chart

```python
plt.figure(figsize=(6,4), dpi=300)

labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

# Alternative way to set hatch
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
plt.show()
```
